```python
dados = dados[dados['DATA'].dt.year == 2023]
```

```python
dados["ID"] = range(1, 126)

dados = dados.query("ID % 3 == 0 or ID == 1")
```

```python
dados = (
    dados.assign(
        DIFF_DAYS = lambda x: x["DATA"].diff().dt.days
    )
)

media_transcorrido = dados["DIFF_DAYS"].std()

media_transcorrido
```
