```python
def prepro(text: str) -> list:
    datalist = nltk.word_tokenize(text)
    stop_words = stopwords.words("portuguese")
    for i in range(len(datalist)):

        datalist[i] = datalist[i].lower()
        datalist[i] = re.sub(r"\W", "", datalist[i])
        datalist[i] = re.sub(r"\s+", " ", datalist[i])
        list = [w for w in datalist if len(w) > 0]
        filtered_sentence = [w for w in list if w.lower() not in stop_words]
    return filtered_sentence
```

```python
def histogram(datalist: list) -> dict:
    wordcount = {}
    for data in datalist:
        words = nltk.word_tokenize(data)
        for word in words:
            if word not in wordcount.keys():
                wordcount[word] = 1

            else:
                wordcount[word] += 1

    return wordcount
```

```python
lista = []
for texto in dados["TEXTOTOTAL"]:
    lista.append(prepro(texto))

lista
```

```python
lista_final = []
for listas in lista:
    for word in listas:
        lista_final.append(word)

lista_final
```

```python
freq = histogram(lista_final)
```

```python
freq_words = heapq.nlargest(50, freq, key = freq.get)

freq_words
```
